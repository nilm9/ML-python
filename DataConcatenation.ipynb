## Dataset join

```python
import pandas as pd
fp = "https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/athletes/"
data = pd.read_csv(fp + "Medals.csv", encoding="ISO-8859-1")
data.head()
```

#### Filter by athlete

```python
a =data["Athlete"].unique().tolist()
len(a)
```

```python
data.shape
```

```python
dataCountry = pd.read_csv(fp + "Athelete_Country_Map.csv", encoding="ISO-8859-1")
```

```python
dataSports = pd.read_csv(fp + "Athelete_Sports_Map.csv", encoding="ISO-8859-1")
```

```python
dataSports.head()
```

```python
dataCountryDp = dataCountry.drop_duplicates(subset="Athlete")
```

```python
dataMainCountry = pd.merge(left=data, right=dataCountryDp, left_on="Athlete", right_on="Athlete")
```

```python
len(dataMainCountry)
```

```python
dataSportsDp = dataSports.drop_duplicates(subset="Athlete")
```

```python
len(dataSports)
```

```python
len(dataSportsDp)
```

```python
dataFinal = pd.merge(left=dataMainCountry, right=dataSportsDp, left_on="Athlete", right_on="Athlete")
```
