**Simulacion de monte carlo





```python
piAvg=0
n=100
n_exp=10
piValueList=[]
for i in range(n_exp):
    value = 0
    x=np.random.uniform(0,1,n).tolist()
    y=np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z=np.sqrt(x[j] * x[j] + y[j] * y[j] )
        if z<=1:
            value +=1
    float_value = float(value)
    piVal=float_value *4/n
    piValueList.append(piVal)
    piAvg += piVal
pi = piAvg/n_exp
print(pi)
plt.plot(piValueList)
```

## Agregation of data from category

```python
import numpy as np
import pandas as pd
gender = ["M","F"]
income= ["Poor","Middle","Rich"]
```

```python
n=500
genderData=[]
incomeData=[]
for i in range(0,500):
    incomeData.append(np.random.choice(income))
    genderData.append(np.random.choice(gender))
```

```python
genderData[:10]
```

```python
#Z -> N(0,1)
#N(m,s) -> m+s*Z
height= 175 +30 * np.random.randn(n)
weight= 75 +25 * np.random.randn(n)
age= 25 +10 * np.random.randn(n)
income= 18000 +3000 * np.random.randn(n)
```

```python
data = pd.DataFrame(
 {
     "Gender": genderData,
     "Economic Status": incomeData,
     "Height": height,
     "Weight" : weight,
     "Age": age,
     "Income": income,
 }
)
```
